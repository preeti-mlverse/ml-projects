# **CREATING DERIVED COLUMNS FROM PAGE INFORMATION**




**Page column - has information about the page name, the main domain, the device type used to access the page, and also the request origin(spider or browser agent),breaking it into multiple columns to use the information to do EDA in these attributes and use it to subset a dataframe based on insights `**

```python
import pandas as pd
import re

# Function to split the string
def split_page_column(text):
    # First split by '.'
    split_by_period = text.split('.')

    # Extract the required parts
    page_name = split_by_period[0]
    domain = split_by_period[1]  # This is typically 'wikipedia' in your example
    device_browser = split_by_period[2]  # This part includes device and browser info

    # Check if the device_browser part contains more than 2 components (i.e., split by '_')
    if len(device_browser.split('_')) >= 3:

        domain_parts = device_browser.split('_')  # Split device_browser into components

        # Construct the main domain and extract device type and browser
        main_domain = domain + "." + domain_parts[0]
        device_type = domain_parts[1]
        browser = domain_parts[2]
        language_origin = page_name.split('_')[-1]  # Extract language origin from page_name
    else:
        # If the device_browser part is not well-formed, set the defaults
        main_domain = np._no_nep50_warning
        device_type = None
        browser = None
        language_origin = None

    # Return the components
    return page_name, language_origin, main_domain, device_type, browser



# Apply the function to the 'Page' column and expand the results into separate columns
df[['Page Name', 'Domain Origin', 'Main Domain', 'Device Type', 'Request Browser']] = df['Page'].apply(split_page_column).apply(pd.Series)


```

# **CHECK FOR STATIONARITY USING DICKEY FULLER TEST**

```python
from statsmodels.tsa.stattools import adfuller
```

```python
# Perform the Augmented Dickey-Fuller (ADF) Test
result = adfuller(final['clicks'])

# Print ADF test results
print("ADF Statistic:", result[0])
print("p-value:", result[1])
print("Critical Values:", result[4])

# Interpret the results
if result[1] < 0.05:
    print("\nThe time series is stationary (reject null hypothesis).")
else:
    print("\nThe time series is not stationary (fail to reject null hypothesis).")
```

**DIFFERENCING TO MAKE THE DATA STATIONARY AS THE DATA IS NON-STATIONARY**

```python
# First differencing
final['clicks_diff'] = final['clicks'].diff()

# Drop NaN values caused by differencing
final.dropna(inplace=True)

```

**The data is stationary after differencing **

**PLOTTING THE ACF AND PACF PLOTS**



```python
# /content/train_1.csv

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot the ACF and PACF
plt.figure(figsize=(12, 4))

# ACF Plot
plt.subplot(1, 2, 1)
plot_acf(final['clicks_diff'], lags=25, ax=plt.gca())  # Adjust lags as needed
plt.title('Autocorrelation Function (ACF)')

# PACF Plot
plt.subplot(1, 2, 2)
plot_pacf(final['clicks_diff'], lags=25, ax=plt.gca(), method='ywm')  # Adjust lags and method as needed
plt.title('Partial Autocorrelation Function (PACF)')

plt.tight_layout()
plt.show()
```

**Conclusion**:

**AUTOCORRELATION OF STATIONARY DATA**

Significant Spike at Lag 1:

The autocorrelation at lag 1 is very strong, with a value close to 1.
This indicates that the current value is highly correlated with the immediate past value.
**No Significant Correlation Beyond Lag 1:**

After lag 1, all spikes fall within the confidence interval (blue shaded region), suggesting no significant correlation with longer lags.
Implications for ACF:
The presence of a single significant spike at lag 1 suggests a Moving Average (MA) component of order q = 1 is appropriate.
The lack of significant correlations at higher lags indicates no further MA terms are needed.

**PARTIAL AUTOCORELATION**

Significant Spike at Lag 1:

The partial autocorrelation is strong and significant at lag 1, with a value close to 1.
This suggests that the immediate past value has a strong direct influence on the current value.
No Significant Correlation Beyond Lag 1:

After lag 1, all spikes fall within the confidence interval, indicating no additional autoregressive relationships.
Implications for PACF:
The single significant spike at lag 1 suggests an Autoregressive (AR) component of order p = 1 is appropriate.
The lack of significant spikes at higher lags suggests no further AR terms are required.

**FINDING THE BEST PARAMETERS FOR ARIMA AND SARIMAX**

```python
import itertools
import statsmodels.api as sm

# Define parameter ranges
p = q = range(0, 5)
d = range(1, 2)
seasonal_p = seasonal_q = range(0, 2)
seasonal_d = range(1, 2)
seasonal_period = [7]

# Grid search for best SARIMAX parameters
best_aic = float("inf")
best_params = None
for param in itertools.product(p, d, q):
    for seasonal_param in itertools.product(seasonal_p, seasonal_d, seasonal_q, seasonal_period):
        try:
            model = sm.tsa.statespace.SARIMAX(train_x,
                                              order=param,
                                              seasonal_order=seasonal_param,
                                              exog=exogenous[:len(train_x)])
            results = model.fit(disp=False)
            if results.aic < best_aic:
                best_aic = results.aic
                best_params = (param, seasonal_param)
        except:
            continue

print(f"Best SARIMAX parameters: {best_params}, AIC: {best_aic}")
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_percentage_error
import warnings
import itertools

# Suppress warnings
warnings.filterwarnings("ignore")

# Define parameter ranges for grid search
p = d = q = range(0, 3)  # Non-seasonal parameters
P = D = Q = range(0, 3)  # Seasonal parameters
m = [7, 12]  # Seasonal periods: weekly (7) or monthly (12)

# Generate all combinations of parameters
param_combinations = list(itertools.product(p, d, q))
seasonal_combinations = list(itertools.product(P, D, Q, m))

# Prepare a dictionary to store the best results for each domain
results = []

# Loop through each domain column
for column in train_x.columns:
    print(f"Processing domain: {column}")

    # Extract the time series
    ts = train_x[column].dropna()

    # Split data into training and testing sets (80% train, 20% test)
    train_size = int(len(ts) * 0.8)
    train, test = ts[:train_size], ts[train_size:]

    best_aic = float("inf")
    best_params = None
    best_model = None

    # Perform grid search for SARIMAX
    for order in param_combinations:
        for seasonal_order in seasonal_combinations:
            try:
                model = SARIMAX(train, order=order, seasonal_order=seasonal_order)
                results_model = model.fit(disp=False)
                aic = results_model.aic

                # Save the best model based on AIC
                if aic < best_aic:
                    best_aic = aic
                    best_params = (order, seasonal_order)
                    best_model = results_model
            except Exception as e:
                # Skip invalid parameter combinations
                continue

    # Forecast with the best model
    if best_model is not None:
        try:
            forecast = best_model.forecast(steps=len(test))
            mape = mean_absolute_percentage_error(test, forecast)
        except Exception as e:
            mape = np.nan
    else:
        mape = np.nan

    # Append results for the current domain
    results.append({
        'Domain': column,
        'Best Order': best_params[0] if best_params else None,
        'Best Seasonal Order': best_params[1] if best_params else None,
        'AIC': best_aic,
        'MAPE': mape
    })

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Display the results
print("SARIMAX Grid Search Results")
print(results_df)
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_percentage_error
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

# Prepare a dictionary to store MAPE values for each domain
results = []

# Loop through each domain column
for column in data.columns:
    print(f"Processing domain: {column}")

    # Extract the time series
    ts = data[column].dropna()

    # Split data into training and testing sets (80% train, 20% test)
    train_size = int(len(ts) * 0.8)
    train, test = ts[:train_size], ts[train_size:]

    # Apply ARIMA
    try:
        arima_model = ARIMA(train, order=(1, 1, 1)).fit()
        arima_forecast = arima_model.forecast(steps=len(test))
        arima_mape = mean_absolute_percentage_error(test, arima_forecast)
    except Exception as e:
        print(f"ARIMA failed for {column}: {e}")
        arima_mape = np.nan

    # Apply SARIMAX
    try:
        sarimax_model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 7)).fit()
        sarimax_forecast = sarimax_model.forecast(steps=len(test))
        sarimax_mape = mean_absolute_percentage_error(test, sarimax_forecast)
    except Exception as e:
        print(f"SARIMAX failed for {column}: {e}")
        sarimax_mape = np.nan

    # Append results
    results.append({
        'Domain': column,
        'ARIMA_MAPE': arima_mape,
        'SARIMAX_MAPE': sarimax_mape
    })

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Display the results
print("ARIMA vs SARIMAX MAPE Comparison")
print(results_df)
```
