**STATSMODEL FOR LINEAR REGRESSION AND VIF**

```python
# Statmodels implementation of Linear regression
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

X = df1[df1.columns.drop('Chance of Admit')]
y = df1["Chance of Admit"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

scaler = StandardScaler()
X_tr_scaled = scaler.fit_transform(X_train)

X_sm = sm.add_constant(X_tr_scaled)  #Statmodels default is without intercept, to add intercept we need to add constant

sm_model = sm.OLS(y_train, X_sm).fit()

print(sm_model.summary())
```

```python
Y_hat = sm_model.predict(X_sm)

MSE_train_set = mean_squared_error(y_train,Y_hat)
print(f"Mean Squared Error: {MSE_train_set}")
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X_t = pd.DataFrame(X_tr_scaled, columns=X_train.columns)
X_t.head()


```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif['Features'] = X_t.columns
vif['VIF'] = [variance_inflation_factor(X_t.values, i) for i in range(X_t.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

**Summary from P-VALUES AND VIF**

As VIF is less than 5, we are not dropping any feature based on VIF.We can check the impact of features using adjusted R-square

Based on P-Values, SOP and LOR are statistically insignificant in determining the chances of admit but we are still not dropping it and we will check the adjusted r-square and decide basis that

# CHECKING ASSUMPTIONS OF LINEAR REGRESSION



1. Multicollinearity check by VIF score  - Drop the features with VIF > 5 one by one
2. The mean of residuals is nearly zero
3. There is linearity of variables such that no pattern in the residual plot
  (Test for Homoscedasticity)
4. Normality of residuals (almost bell-shaped curve in residuals distribution, points in QQ plot are almost all on the line)


```python
Y_hat = sm_model.predict(X_sm)
```

```python
errors = y_train - Y_hat
```

ASSUMPTION 1 -
LINEARITY BETWEEN DEPENDENT AND INDEPENDENT VARIABLES

```python
## LINEARITY CHECK

# there are some ordinal categorical variables and ordinal variables that do not have direct linearity
sns.pairplot(df1.drop(['SOP','LOR'],axis=1), y_vars=["Chance of Admit"]);
```

```python
plt.scatter(Y_hat, errors)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs Predicted")
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

There is no pattern in residual plot

ASSUMPTION 2: CHECKING MULTICOLINEARITY USING VIF

```python
vif = pd.DataFrame()
X_t = pd.DataFrame(X_tr_scaled, columns=X_train.columns)
X_t.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif['Features'] = X_t.columns
vif['VIF'] = [variance_inflation_factor(X_t.values, i) for i in range(X_t.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

BASED ON THE VALUES OF VIF, all are < 5, There is no multicollinearity, so the assumption is validated

ASSUMPTION 3 - CHECKING FOR HETEROSKEDASTICITY IN VARIANCE OF RESIDUALS WITH THE PREDICTED VALUES




```python
# Plot residuals vs predicted values (same as the second part of Linearity check)
plt.scatter(Y_hat, errors)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs Predicted")
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
# Performing the Goldfeld-Quandt test to check for Homoscedasticity -
from statsmodels.compat import lzip
import statsmodels.stats.api as sms

name = ['F statistic', 'p-value']
test = sms.het_goldfeldquandt(y_train, X_sm)
lzip(name, test)


# h0: it's homo
# ha: it's not
```

Based on p-value and f-statistic, we cannot conclude about heteroskedasticity

P-value is > 0.05(alpha) which implies that null hypothesis could not be rejected.There is homoskedasticity - the variance is almost constant through the predicted values

ASSUMPTION 4: NORMAL DISTRIBUTION OF ERRORS

```python
#### checking for normal distribution of errors/residuals - mean is centered close to zero

import seaborn as sns
sns.histplot(errors, kde=True)
plt.xlabel("Residuals")
plt.title("Histogram of residuals")
```

```python
# Q-Q plot'
import scipy.stats as stats

stats.probplot(errors, dist="norm", plot=plt)
plt.show()
```

Residuals mean is nearly zero
